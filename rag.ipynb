```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
```

```python
pages = PyPDFLoader('//content/pdfllm.pdf')
```

```python
print(pages)
```

```python
page = pages.load_and_split()
```

```python
#  this is the class Recursive
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1024, #parameter in the class recursive
    chunk_overlap=64
)
```

```python
print(f"Total chunks: {len(page)}")
print(f"\nFirst chunk:\n{page[0].page_content}")
print(f"\nSecond chunk:\n{page[1].page_content}")
```

```python
Embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
    model_kwargs={'device': 'cuda'}
)

vector_db = FAISS.from_documents(page, Embeddings)
```

```python
question ="""
what is purpose of decoder
"""


results = vector_db.similarity_search(question,k=2)

results
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
import torch
tokenizer = AutoTokenizer.from_pretrained("mistralai/Mistral-7B-Instruct-v0.2")
model = AutoModelForCausalLM.from_pretrained("mistralai/Mistral-7B-Instruct-v0.2", quantization_config=bnb_config)
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_core.prompts import ChatPromptTemplate
from langchain_community.llms import HuggingFacePipeline
from langchain_classic.chains import RetrievalQA
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig, pipeline
import torch
import os
import warnings

warnings.filterwarnings('ignore')
```

```python
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=64  # ← fix this
)
lc_pipeline = HuggingFacePipeline(pipeline=pipe)

qa_chain = RetrievalQA.from_chain_type(
    llm=lc_pipeline,
    retriever=vector_db.as_retriever(search_kwargs={'k': 3}),
    return_source_documents=True,
    chain_type_kwargs={'prompt': prompt}
)
```

```python
question ="""
what is purpose of decoder """
```

```python
result = qa_chain({"query":question})
```

```python
# Safe way to print result
output = result["result"]

if "Answer:" in output:
    print(output.split("Answer:")[1].strip())
else:
    print(output)
```
